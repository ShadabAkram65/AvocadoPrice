# Data Preparation and Cleaning

#### Drop the 'Unnamed: 0' column

```python
df.drop('Unnamed: 0',axis='columns',inplace=True)
```

```python
df
```

#### Modifying the columns names

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(" ","")

df.rename(columns = {'type':'Type','year':'Year','region':'Region'},inplace=True)
df
```

#### Check Outliers

* AveragePrice

```python
fig = px.histogram(df['AveragePrice'],title='Check the average price outliers')
fig.show()
```

There are some outliers <br>
So because column 'AveragePrice' is positively skewed distribution, so I will use IQR

```python
# Finding the IQR

q1 = df['AveragePrice'].quantile(0.25)

q3 = df['AveragePrice'].quantile(0.75)

#The value of IQR
IQR = q3 - q1
```

```python
upper_limit = q3 + (1.5) * IQR

lower_limit = q1 - (1.5) * IQR
```

#### Capping Technique

```python
df['AveragePrice']=pd.DataFrame(np.where(df['AveragePrice'] > upper_limit,upper_limit,
                   (np.where(df['AveragePrice'] < lower_limit,lower_limit, df['AveragePrice']))),columns=['AveragePrice'])
```

```python
# Average Price after capping
fig = px.histogram(df['AveragePrice'], title='Average price after capping')

fig.show()
```

* TotalVolume

```python

fig = px.histogram(df['TotalVolume'], title="Check TotalVolume outliers")
fig.show()
```

There are some outliers in this section.<br>
And it is also positively skewed distrubution

```python
# Finding the IQR
q1 = df['TotalVolume'].quantile(0.25)

q3 = df['TotalVolume'].quantile(0.75)

# The value of IQR
IQR = q3 - q1
```

```python
upper_limit = q3 + (1.5) * IQR

lower_limit = q1 - (1.5) * IQR
```

#### Capping Technique

```python
df['TotalVolume'] = pd.DataFrame(np.where(df['TotalVolume'] > upper_limit,upper_limit,
                  (np.where(df['TotalVolume'] < lower_limit,lower_limit,df['TotalVolume']))), columns=['TotalVolume'])
```

```python
# TotalVolume after capping
fig = px.histogram(df['TotalVolume'],title='Check TotalVolume outliers')

fig.show()
```

#### Change DataType of 'Date' and 'Year' columns

```python
df['Year'] = pd.to_datetime(df['Year'],format='%Y')
df['Date'] = pd.to_datetime(df['Date'])

df.info()
```

# EDA (Exploratory Data Analysis)

#### Q1: How has the 'AveragePrice' change over the time?

```python
avg_price_time=df.groupby('Date')['AveragePrice'].mean().reset_index()

avg_price_time
```

```python
fig = px.line(avg_price_time,x='Date',y='AveragePrice',
      title='Average Price Changed Over the Time', labels=dict(AveragePrice='Average Price'))
style_tick_marks()
fig.show()
```

#### Q2:What is the correlation between 'TotalVolume' and 'AveragePrice'?

```python
df['TotalVolume'].corr(df['AveragePrice'])
```

```python
fig=px.scatter(df,x='TotalVolume',y='AveragePrice',
    title='Correlation between TotalVolume and AveragePrice',labels=dict(TotalVolume='Total Volume', AveragePrice='Average Price'))
style_tick_marks()
fig.show()
```

#### Q3:How does the 'AveragePrice' vary by 'Type' of Avocado?

```python
avg_by_type=df.groupby('Type')['AveragePrice'].mean().reset_index()
avg_by_type
```

```python
fig=px.pie(avg_by_type,values='AveragePrice',names='Type', title='Average Price Vary by Time')
fig.show()
```

##### This is generally expected. Organic farming practices are often more costly due to stricter regulations and lower yields 

#### Q4:How does the 'AveragePrice' vary across the different 'Regions'? 

```python
avg_by_regions=df.groupby('Region')['AveragePrice'].mean().reset_index().sort_values(by='AveragePrice',ascending=False)

avg_by_regions
```

```python
fig=px.bar(avg_by_regions, x='Region',y='AveragePrice',
    title="Average Price of Avocado in Different Regions", labels=dict(AveragePrice='Average Price'))

fig.show()
```

#### Q5:How has the sales volume 'TotalVolume' changed year over year?

```python
total_volume_time = df.groupby('Year')['TotalVolume'].mean().reset_index()

total_volume_time
```

```python
fig=px.line(total_volume_time,x='Year',y='TotalVolume',
    title='Total Volume changed over the Years',labels=dict(TotalVolume='Total Volume'))
style_tick_marks()
fig.show()
```

#### Q6:Are there any seasonal trends in avocado prices or sales volumes?

```python
df['Month']=df['Date'].dt.month
```

```python
df_price = df.groupby('Month')['AveragePrice'].mean().reset_index()

df_price
```

```python
df_volume = df.groupby('Month')['TotalVolume'].mean().reset_index()

df_volume
```

```python
fig=px.line(df_price,x='Month',y='AveragePrice', title= 'Average Price by Month',
    labels=dict(AveragePrice='Average Price'))
style_tick_marks()
fig.show()
```

```python
fig=px.line(df_volume,x='Month',y='TotalVolume',title='Total Volume by Month',
    labels=dict(TotalVolume='Total Volume'))
style_tick_marks()
fig.show()
```
